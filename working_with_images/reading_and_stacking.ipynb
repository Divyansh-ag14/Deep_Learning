## Loading a RGB image

```python
# reading the image using its path
image = imread('dog.jpg')
```

```python
#checking shape of image
image.shape
```

```python
# image matrix
image
```

## Printing different channels of RGB image

```python
# plotting the image
imshow(image)
```

```python
#printing only the red channel
red_filter = [1,0,0]
imshow(image*red_filter)
```

```python
# printing only the green channel
green_filter = [0,1,0]
imshow(image*green_filter)
```

```python
# individual channel using cmap
imshow(image[:,:,0], cmap='Reds')
```

## Loading MRI image

```python
# loading library to read images
import nibabel as nib
```

```python
image_mri = nib.load('sM00223_002.hdr')
image_mri = image_mri.get_data()
```

```python
image_mri.shape
```

```python
image_mri = image_mri.squeeze()
image_mri.shape
```

```python
# visualizing images
import matplotlib.pyplot as plt

fig,ax = plt.subplots(nrows=2,ncols=3, figsize=(20,20))

ax[0,0].imshow(image_mri[:, :, 0], cmap = 'gray')
ax[0,1].imshow(image_mri[:, :, 11], cmap = 'gray')
ax[0,2].imshow(image_mri[:, :, 23], cmap = 'gray')
ax[1,0].imshow(image_mri[:, :, 37], cmap = 'gray')
ax[1,1].imshow(image_mri[:, :, 42], cmap = 'gray')
ax[1,2].imshow(image_mri[:, :, 51], cmap = 'gray')
```

## Loading Multiple Images

```python
# reading multiple images
from glob import glob
image_name = glob("*.jpg")
```

```python
image_name
```

```python
image_=[]
for i in image_name:
  image = imread(i)
  imshow(image)
  plt.show()
  image_.append(image)

final_images = np.array(image_)
```

```python
# checking the shape
final_images.shape
```

```python
final_images[0].shape, final_images[1].shape, final_images[2].shape
```

## Images of different size

```python
# reshape images
final_resized = []
for i in range (0,3):
  temp = resize(final_images[i], (340, 450))
  final_resized.append(temp)

final_resized = np.array(final_resized)
```

```python
# print shape
final_resized.shape
```

```python
# visualizing images
import matplotlib.pyplot as plt

fig,ax = plt.subplots(nrows=1, ncols=2)

ax[0].imshow(final_resized[2])
ax[1].imshow(final_images[0])
```

```python
imsave('resized_car.jpg',final_resized[2])
```
